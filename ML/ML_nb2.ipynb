```python
import pandas as pd
from google.colab import files
uploaded=files.upload()
df=pd.read_csv("spam.csv")
df
```

```python
df.groupby('Category').describe()
```

```python
df['spam']=df['Category'].apply(lambda x: 1 if x=='spam' else 0)
df.head()
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(df.Message,df.spam,test_size=0.2)
```

```python
 from sklearn.feature_extraction.text import CountVectorizer
 cv=CountVectorizer()
 xtraincount=cv.fit_transform(xtrain.values)
 xtraincount.toarray()[:3]
```

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(xtraincount,ytrain)
```

```python
emails = [
    'Hey mohan, can we get together to watch footbal game tomorrow?',
    'Upto 20% discount on parking, exclusive offer just for you. Dont miss this reward!'
]
emailscount=cv.transform(emails)
model.predict(emailscount)
```

```python
xtestcount=cv.transform(xtest)
model.score(xtestcount,ytest)
```

```python
from sklearn.pipeline import Pipeline
clf=Pipeline([
    ('vectorizer',CountVectorizer()),
    ('nb',MultinomialNB())
])
```

```python
clf.fit(xtrain,ytrain)
clf.score(xtest,ytest)
```
