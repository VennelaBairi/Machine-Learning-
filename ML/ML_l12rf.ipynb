```python
cols_to_use = ['Suburb', 'Rooms', 'Type', 'Method', 'SellerG', 'Regionname', 'Propertycount',
               'Distance', 'CouncilArea', 'Bedroom2', 'Bathroom', 'Car', 'Landsize', 'BuildingArea', 'Price']
df = df[cols_to_use]
```

```python
len(cols_to_use)
```

```python
df.isna().sum()
```

```python
with_0=['Propertycount','Bedroom2','Car','Distance']
df[with_0]=df[with_0].fillna(0)
df.Bathroom=df.Bathroom.fillna(1)
df.isna().sum()
```

```python
df['Landsize'] = df['Landsize'].fillna(df.Landsize.mean())
df['BuildingArea'] = df['BuildingArea'].fillna(df.BuildingArea.mean())
df.isna().sum()
```

```python
df.dropna(inplace=True)
df.isna().sum()
```

```python
df.shape
```

```python
#df.dropna(subset=['Regionname', 'CouncilArea'], inplace=True)
```

```python
df.shape
```

```python
df=pd.get_dummies(df,drop_first=True)
df.head()
```

```python
df.columns
```

```python
x=df.drop('Price',axis='columns')
y=df.Price
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.3, random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(xtrain, ytrain)
```

```python
reg.score(xtest,ytest)
```

```python
from sklearn.linear_model import Lasso
lasso_reg=Lasso(alpha=50,max_iter=100,tol=0.1)
lasso_reg.fit(xtrain, ytrain)
```

```python
lasso_reg.score(xtest,ytest)
```

```python
from sklearn.linear_model import Ridge
ridge_reg=Ridge(alpha=50,max_iter=100,tol=0.1)
ridge_reg.fit(xtrain, ytrain)
```

```python
ridge_reg.score(xtest,ytest)
```
