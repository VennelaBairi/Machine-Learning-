```python
import pandas as pd
from sklearn.datasets import load_iris
iris=load_iris()
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df['target']=iris.target
df['flower']=df.target.apply(lambda x: iris.target_names[x])
df
```

```python
iris.target_names
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(df.drop(['target', 'flower'], axis='columns'), df['target'], test_size=0.2)
```

```python
from sklearn.svm import SVC
model=SVC()
model.fit(xtrain,ytrain)
model.score(xtest,ytest)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(SVC(kernel='linear',C=10),iris.data,iris.target,cv=5)
```

```python
cross_val_score(SVC(kernel='rbf',C=10),iris.data,iris.target,cv=5)
```

```python
import numpy as np
kernels=['rbf','linear']
Cs=[1,10,20]
gammas=['auto','scale']
avg_scores={}
for kval in kernels:
  for cval in Cs:
    for gval in gammas:
      score=cross_val_score(SVC(kernel=kval,C=cval,gamma=gval),iris.data,iris.target,cv=5)
      avg_scores[kval+'_'+str(cval)+'_'+gval]=np.average(score)

avg_scores
```

```python
from sklearn.model_selection import GridSearchCV

clf=GridSearchCV(
    SVC(),
    {
        'kernel':['rbf','linear'],
        'C':[1,10,20],
        'gamma':['auto','scale']
    },cv=5,return_train_score=False
)
clf.fit(iris.data,iris.target)
```

```python
df=pd.DataFrame(clf.cv_results_)
df
```

```python
df[['params','mean_test_score']]
```

```python
from sklearn.model_selection import RandomizedSearchCV
rs=RandomizedSearchCV(
    SVC(),
    {
        'kernel':['rbf','linear'],
        'C':[1,10,20],
        'gamma':['auto','scale']
    },cv=5,return_train_score=False,n_iter=2
)
rs.fit(iris.data,iris.target)
df=pd.DataFrame(rs.cv_results_)[['param_C','param_kernel','param_gamma','mean_test_score']]
df
```
