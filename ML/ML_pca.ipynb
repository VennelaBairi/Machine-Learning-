```python
import pandas as pd
from sklearn.datasets import load_digits
ds=load_digits()
ds.keys()
```

```python
ds.data[0]
```

```python
ds.data[0].reshape(8,8)
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
plt.gray()
plt.matshow(ds.data[0].reshape(8,8))
```

```python
ds.target.shape
```

```python
df=pd.DataFrame(ds.data,columns=ds.feature_names)
df.head()
```

```python
df.describe()
```

```python
x=df
y=ds.target
```

```python
x.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
xscaled=scaler.fit_transform(x)
xscaled
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(xscaled, y, test_size=0.2,random_state=30)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(xtrain,ytrain)
model.score(xtest,ytest)
```

```python
from sklearn.decomposition import PCA
pca=PCA(0.95)
x_pca=pca.fit_transform(x)
x_pca.shape
```

```python
pca.n_components_
```

```python
pca.explained_variance_ratio_
```

```python
xtrain_pca, xtest_pca, ytrain, ytest = train_test_split(x_pca, y, test_size=0.2,random_state=30)
```

```python
model=LogisticRegression(max_iter=1000)
model.fit(xtrain_pca,ytrain)
model.score(xtest_pca,ytest)
```
